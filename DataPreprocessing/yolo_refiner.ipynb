```python
import os
import shutil
base_path = 'images'

def list_folders_in_path(path):
    folders = [name for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))]
    return folders


def list_png_files_in_path(path):
    png_files = [f for f in os.listdir(path) if f.endswith('.png') and os.path.isfile(os.path.join(path, f))]
    return png_files

dest_dir = 'train_yolo'


def copy_and_rename_file(src_file, dest_dir, new_name):
    
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    dest_file = os.path.join(dest_dir, new_name)
    shutil.copy2(src_file, dest_file)

sum = 0
folders_1 = list_folders_in_path(base_path)
print(folders_1)
for folder in folders_1:
    base_path_1 = base_path + "/" + folder
    folders_2 = list_folders_in_path(base_path_1)
    #print(folders_2)
    for folder2 in folders_2:
        base_path_2 = base_path_1 + "/"+ folder2
        pngs = list_png_files_in_path(base_path_2)
        #print(pngs)
        sum += len(pngs)
        for png in pngs:
            png_path = base_path_2+"/"+png
            #print(base_path_2+"/"+png)
            
            new_name = folder2 + "_" + png
            print(new_name)
            copy_and_rename_file(png_path, dest_dir, new_name)

print(sum)
```
